- 将经过审查的文本（即经过上述替换处理的文本）保存到一个新文件（该文件路径通过函数参数`outfile`设定）

````python
# your code here
def censorTextfile(infile, outfile):
    blockstring=input('请输入需要屏蔽的英文单词')
    blocklist=blockstring.split(' ')
    with open(infile,'r') as f1:
        new_lines=[]
        for line in f1.readlines() :
            List=[]
            for word in line.strip().split():
                if word in blocklist:
                    for char in word:
                        list_=[]
                        list_.append('%d'%(ord(char)))
                        newword=''.join(list_)
                        List.append(newword)
                else:
                    List.append(word)
            new_line=' '.join(List)
            new_lines.append(new_line)

        
    with open(outfile,'w') as f2:
        f2.write('\n'.join(new_lines))

censorTextfile('data/text.txt','data/newtext.txt')



````

### Q-Bonus. 函数编写与数据结构的使用：判断括号使用是否合法（10pts，该次作业允许最终总分高于100分，但是4次作业总分上限不能高于400分）
Step 1: 编写函数`is_valid_parenthese_str(str1)`, 该函数接受一个字符串参数`str1`, 假设该字符串只包含括号（即 `(`, `)`, `{`, `}`, `[`, `]` ）。函数需要判断这些括号是否按正确的顺序闭合。举例：`()`和`()[]{}`和`[()]`是合法的，而 `[)`, `({[)]`和 `{{{`是不合法的

Step 2: 用以下几个字符串测试你的函数：
  * `[{}([])] -> True`
  * `(){}[] -> True`
  * `()[{)} -> False`

提示：想象你在进行一个消消乐游戏，从左往右，碰到左括号就保存到一个待消除左括号列表里，碰到右括号就和最近的左括号消除。如果最后待消除左括号列表为空，那么括号合法。以下已经给出了函数的框架，你只需要补充完整即可。如果你的思路不一样，也可以重新写一个函数。


````python
# your codes here
def is_valid_parenthese_str(str1):
    '''
    判断输入的字符串是否是有效的括号字符串
    '''
    recent_left_parenthese = []
    pmap = {"(": ")", 
            "{": "}", 
            "[": "]"}
    for parenthese in str1:
        if parenthese in pmap: # 碰到新的左括号
           recent_left_parenthese.append(parenthese) 
        else: # 碰到右括号
            if pmap[recent_left_parenthese[-1]]==parenthese:
                del recent_left_parenthese[-1]
            else:
                break
    if  not recent_left_parenthese:
        return 'Ture'
    else:
        return 'False'

    
print(is_valid_parenthese_str("[{}([])]"))
print(is_valid_parenthese_str("(){}[]"))
print(is_valid_parenthese_str("()[{)}"))
````

````python
#答案
def is_valid_parenthese_str(str1):
    '''
    判断输入的字符串是否是有效的括号字符串
    '''
    recent_left_parenthese = []
    pmap = {"(": ")", 
            "{": "}", 
            "[": "]"}
    for parenthese in str1:
        if parenthese in pmap: # 碰到新的左括号
            recent_left_parenthese.append(parenthese)
        else: # 碰到右括号
            if len(recent_left_parenthese) == 0:
                return False # 如果左边没有左括号
            most_recent_left_parenthese = recent_left_parenthese.pop() # 删除最后一个元素并返回该元素
            if pmap[most_recent_left_parenthese] != parenthese:
                return False
    return len(recent_left_parenthese) == 0  # 第一个左括号必须有合适的右括号

print(is_valid_parenthese_str("[{}([])]"))
print(is_valid_parenthese_str("(){}[]"))
print(is_valid_parenthese_str("()[{)}"))
````
