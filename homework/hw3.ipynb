  * `[{}([])] -> True`

  * `()[{)} -> False`

æç¤ºï¼šæƒ³è±¡ä½ åœ¨è¿›è¡Œä¸€ä¸ªæ¶ˆæ¶ˆä¹æ¸¸æˆï¼Œä»å·¦å¾€å³ï¼Œç¢°åˆ°å·¦æ‹¬å·å°±ä¿å­˜åˆ°ä¸€ä¸ªå¾…æ¶ˆé™¤å·¦æ‹¬å·åˆ—è¡¨é‡Œï¼Œç¢°åˆ°å³æ‹¬å·å°±å’Œæœ€è¿‘çš„å·¦æ‹¬å·æ¶ˆé™¤ã€‚å¦‚æœæœ€åå¾…æ¶ˆé™¤å·¦æ‹¬å·åˆ—è¡¨ä¸ºç©ºï¼Œé‚£ä¹ˆæ‹¬å·åˆæ³•ã€‚ä»¥ä¸‹å·²ç»ç»™å‡ºäº†å‡½æ•°çš„æ¡†æ¶ï¼Œä½ åªéœ€è¦è¡¥å……å®Œæ•´å³å¯ã€‚å¦‚æœä½ çš„æ€è·¯ä¸ä¸€æ ·ï¼Œä¹Ÿå¯ä»¥é‡æ–°å†™ä¸€ä¸ªå‡½æ•°ã€‚


````python
# your codes here
def is_valid_parenthese_str(str1):
    '''
    åˆ¤æ–­è¾“å…¥çš„å­—ç¬¦ä¸²æ˜¯å¦æ˜¯æœ‰æ•ˆçš„æ‹¬å·å­—ç¬¦ä¸²
    '''
    recent_left_parenthese = []
    pmap = {"(": ")", 
            "{": "}", 
            "[": "]"}
    for parenthese in str1:
        if parenthese in pmap: # ç¢°åˆ°æ–°çš„å·¦æ‹¬å·
           recent_left_parenthese.append(parenthese) 
        else: # ç¢°åˆ°å³æ‹¬å·
            if pmap[recent_left_parenthese[-1]]==parenthese:
                del recent_left_parenthese[-1]
            else:
                break
    if  not recent_left_parenthese:
        return 'Ture'
    else:
        return 'False'

    
print(is_valid_parenthese_str("[{}([])]"))
print(is_valid_parenthese_str("(){}[]"))
print(is_valid_parenthese_str("()[{)}"))
````

````python
#ç­”æ¡ˆ
def is_valid_parenthese_str(str1):
    '''
    åˆ¤æ–­è¾“å…¥çš„å­—ç¬¦ä¸²æ˜¯å¦æ˜¯æœ‰æ•ˆçš„æ‹¬å·å­—ç¬¦ä¸²
    '''
    recent_left_parenthese = []
    pmap = {"(": ")", 
            "{": "}", 
            "[": "]"}
    for parenthese in str1:
        if parenthese in pmap: # ç¢°åˆ°æ–°çš„å·¦æ‹¬å·
            recent_left_parenthese.append(parenthese)
        else: # ç¢°åˆ°å³æ‹¬å·
            if len(recent_left_parenthese) == 0:
                return False # å¦‚æœå·¦è¾¹æ²¡æœ‰å·¦æ‹¬å·
            most_recent_left_parenthese = recent_left_parenthese.pop() # åˆ é™¤æœ€åä¸€ä¸ªå…ƒç´ å¹¶è¿”å›è¯¥å…ƒç´ 
            if pmap[most_recent_left_parenthese] != parenthese:
                return False
    return len(recent_left_parenthese) == 0  # ç¬¬ä¸€ä¸ªå·¦æ‹¬å·å¿…é¡»æœ‰åˆé€‚çš„å³æ‹¬å·

print(is_valid_parenthese_str("[{}([])]"))
print(is_valid_parenthese_str("(){}[]"))
print(is_valid_parenthese_str("()[{)}"))
````
